# 01.Wykres liniowy

```
import matplotlib
matplotlib.__version__
```

```
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
sns.set()
```

```
# tworzenie obiektu typu figure
fig = plt.figure()
fig.show()
```

```
# tworzenie linii prostej
plt.plot([0, 1], [0, 3])
# aby miec sam wykres trzeba go przypisac do zmiennej
```

```
plt.plot([0, 1], [0, 3], label = 'line', color = 'green')
# nazw osi x
plt.xlabel('oś x')
plt.ylabel('oś y')
plt.title('Wykres 1')
# aby była legenda
plt.legend()
```

```
plt.plot([0, 1, 2], [0, 3, 1], label = 'line', color = 'green')
# nazw osi x
plt.xlabel('oś x')
plt.ylabel('oś y')
plt.title('Wykres 1')
# aby była legenda
plt.legend()
```

```
# wykres sinusoidalny
x = np.arange(-3, 3, 0.1)
x_sin = np.sin(x)
x_cos = np.cos(x)
# podajemy wartosci x ów oraz wartości y 
plt.plot(x, x_sin)
# gdy dodamy jeszcze jedno polecenie domyślnie dwa wykresy są nałożone na siebie
plt.plot(x, x_cos)
```

```
# tworzenie wielu wykresów
#w tym przypadku jedynka okresla liczbe wierszy
# dwójka liczbe kolumn, ktore chcemy uzyskac
# kolejna jedynka odnosi sie do konkretnego wykresu
plt.subplot(121)
plt.plot([0,1,2], [3, 2, 5])
# aby stworzyc wykres obok trzeba wpisac 2 na koncu subplot
# czerwona linia przerywana r --
plt.subplot(122)
plt.plot([0,1,2], [3, 4, 3], 'r--')
```

```
# wielkość wykresu
fig = plt.figure(figsize=(10, 6))
plt.subplot(131)
plt.plot([0,1,2], [3, 2, 5])
plt.subplot(132)
plt.plot([0,1,2], [3, 4, 3], 'r--')
plt.subplot(133)
plt.plot([0,1,2], [3, 4, 3], 'ro')
# w spyderze trzeba fig.show()
```

# 02.Wykres słupkowy

```
# tworzenie wykresu słupkowego
plt.bar(x = [0, 1, 2, 3], height = [2.2, 3.2, 1.2, 1.1])
```

```
plt.bar(x = ['a', 'b', 'c', 'd'], height = [2.2, 3.2, 1.2, 1.1])
```

```
# wykres horyzontalny
plt.barh(y = ['a', 'b', 'c', 'd'], width = [2.2, 3.2, 1.2, 1.1])
```
