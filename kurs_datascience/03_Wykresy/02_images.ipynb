```
import matplotlib.pyplot as plt
import matplotlib.image as mpimg
```

```
img = mpimg.imread('office.jpg')
img
```

```
img.shape
```

```
# wyswietlanie obrazu
plt.imshow(img)
#wylaczanie osi
plt.axis('off')
```

```
# wyświetlanie kilku obrazów
fig = plt.figure(figsize =(12, 10))
plt.subplot(221)
plt.imshow(mpimg.imread('office.jpg'))
plt.subplot(222)
plt.imshow(mpimg.imread('bussines.jpg'))
plt.subplot(223)
plt.imshow(mpimg.imread('children.jpg'))
plt.subplot(224)
plt.imshow(mpimg.imread('notebook.jpg'))
```

```
# pętla
fig = plt.figure(figsize =(12, 10))

for idx, img in enumerate(['office.jpg', 'bussines.jpg', 'children.jpg',
                           'notebook.jpg']):
    plt.subplot(220 + idx + 1 )
    plt.imshow(mpimg.imread(img))
```

```
# zmiana rozmiwaru siatki
fig = plt.figure(figsize =(12, 10))

for idx, img in enumerate(['office.jpg', 'bussines.jpg', 'children.jpg',
                           'notebook.jpg']):
    plt.subplot(140 + idx + 1 )
    plt.imshow(mpimg.imread(img))
```
